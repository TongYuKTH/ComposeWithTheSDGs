```python
import pandas as pd
import os
from pathlib import Path

# 设置文件夹路径
folder_path = os.getcwd()

# 初始化列表存储所有数据
all_data = []


for i in range(1, 18):  # 1 到 17
    filename = f'sdg{i}.csv'
    filepath = os.path.join(folder_path, filename)
    
    # 检查文件是否存在
    if os.path.exists(filepath):
        try:
            df = pd.read_csv(filepath, encoding='utf-8-sig')
            
            # 验证必需的列
            required_cols = ['Country', 'Year', 'SDG', 'Value']
            if all(col in df.columns for col in required_cols):
                all_data.append(df)
                print(f"✓ SDG{i:2d}: {len(df):5d} 行  ({df['Country'].nunique()} 个国家)")
            else:
                print(f"✗ SDG{i:2d}: 缺少必需的列 {required_cols}")
        except Exception as e:
            print(f"✗ SDG{i:2d}: 读取失败 - {e}")
    else:
        print(f"⊘ SDG{i:2d}: 文件不存在")

print("="*60)

# 合并所有数据
if all_data:
    df_merged = pd.concat(all_data, ignore_index=True)
    
    # 排序：按国家、年份、SDG
    df_merged = df_merged.sort_values(['Country', 'Year', 'SDG'])
    
    # 重置索引
    df_merged = df_merged.reset_index(drop=True)
    
    # 保存合并后的文件
    output_file = os.path.join(folder_path, 'all_sdg_data.csv')
    df_merged.to_csv(output_file, index=False, encoding='utf-8-sig')
    
   
    print(f" 总数据统计:")
    print(f"   总行数: {len(df_merged)}")
    print(f"   国家数: {df_merged['Country'].nunique()}")
    print(f"   SDG数量: {df_merged['SDG'].nunique()}")
    print(f"   年份范围: {df_merged['Year'].min()} - {df_merged['Year'].max()}")
    
    print(f"\n 每个SDG的数据量:")
    sdg_counts = df_merged['SDG'].value_counts().sort_index()
    for sdg, count in sdg_counts.items():
        print(f"   SDG{sdg:2d}: {count:5d} 行")
    
    print(f"\n 输出文件: {output_file}")
    
    print(f"\n 数据预览:")
    print(df_merged.head(10))
    
else:
    print("\n 没有找到任何有效的SDG文件！")
```
