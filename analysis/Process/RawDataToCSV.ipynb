## SDG1 
取BOTHSEX,'2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'

```python
import pandas as pd

# 读取Excel
df = pd.read_excel('SDG1.xlsx', sheet_name=0)

# 只保留BOTHSEX
df = df[df['Sex'] == 'BOTHSEX']

# 选择需要的列：国家名和年份列
years = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
df_long = pd.melt(
    df,
    id_vars=['GeoAreaName'],
    value_vars=years,
    var_name='Year',
    value_name='Value'
)

# 删除空值
df_long = df_long.dropna(subset=['Value'])

# 重命名列
df_long = df_long.rename(columns={'GeoAreaName': 'Country'})

# 添加SDG编号
df_long['SDG'] = 1

# 调整列顺序
df_long = df_long[['Country', 'Year', 'SDG', 'Value']]

# 保存（使用UTF-8编码处理特殊字符）
df_long.to_csv('sdg1.csv', index=False, encoding='utf-8-sig')

print(f"完成！共 {len(df_long)} 行数据")
print(df_long.head(10))
```

## SDG4
先取BOTHSEX,ALLAREA,Quantile-_T然后对同一年多条数据取平均

```python
import pandas as pd

# 读取Excel
df = pd.read_excel('SDG4.xlsx', sheet_name=0)

# 精确筛选：
# - Sex = BOTHSEX (不分性别)
# - Location = ALLAREA (全国总体，不分城乡)
# - Quantile = _T (所有收入层级)
# - Nature = C (完成率，而不是N=净入学率)
df = df[
    (df['Sex'] == 'BOTHSEX') & 
    (df['Location'] == 'ALLAREA') &
    (df['Quantile'] == '_T') &
    (df['Nature'] == 'C')
]

# 选择需要的列
df = df[['GeoAreaName', 'TimePeriod', 'Value']]

# 删除空值
df = df.dropna(subset=['Value'])

# 按国家和年份分组，对三个教育层级取平均
df = df.groupby(['GeoAreaName', 'TimePeriod'], as_index=False)['Value'].mean()

# 重命名列
df = df.rename(columns={
    'GeoAreaName': 'Country',
    'TimePeriod': 'Year'
})

# 添加SDG编号
df['SDG'] = 4

# 调整列顺序
df = df[['Country', 'Year', 'SDG', 'Value']]

# 保存
df.to_csv('sdg4.csv', index=False, encoding='utf-8-sig')

print(f"完成！共 {len(df)} 行数据")
print(df.head(10))
```

## SDG11 (df['Sex'] == 'BOTHSEX') & 
    (df['Age'] == 'ALLAGE') &
    (df['Disability status'] == '_T')
] 然后按城市取平均

```python
import pandas as pd

# 读取Excel
df = pd.read_excel('SDG11.xlsx', sheet_name=0)

# 筛选维度：
# - Sex = BOTHSEX (不分性别)
# - Age = ALLAGE (所有年龄)
# - Disability status = _T (不分残疾状况)
df = df[
    (df['Sex'] == 'BOTHSEX') & 
    (df['Age'] == 'ALLAGE') &
    (df['Disability status'] == '_T')
]

# 确定年份列
years = [str(year) for year in range(2000, 2025)]

# 转换为长格式
df_long = pd.melt(
    df,
    id_vars=['GeoAreaName', 'Cities'],
    value_vars=years,
    var_name='Year',
    value_name='Value'
)

# 删除空值
df_long = df_long.dropna(subset=['Value'])

# 按国家和年份聚合，取该国所有城市的平均值
df_country = df_long.groupby(['GeoAreaName', 'Year'], as_index=False)['Value'].mean()

# 重命名列
df_country = df_country.rename(columns={'GeoAreaName': 'Country'})

# 添加SDG编号
df_country['SDG'] = 11

# 调整列顺序
df_country = df_country[['Country', 'Year', 'SDG', 'Value']]

# 保存
df_country.to_csv('sdg11.csv', index=False, encoding='utf-8-sig')

print(f"完成！共 {len(df_country)} 行数据")
print(df_country.head(10))
```

## SDG16

```python
import pandas as pd
import numpy as np

# 读取Excel
df = pd.read_excel('SDG16.xlsx', sheet_name=0)

# 确定年份列
years = [str(year) for year in range(2009, 2025)]

# 转换为长格式
df_long = pd.melt(
    df,
    id_vars=['GeoAreaName', 'Sex'],
    value_vars=years,
    var_name='Year',
    value_name='Value'
)

# 删除空值
df_long = df_long.dropna(subset=['Value'])


# 智能选择性别数据
result_list = []

for country in df_long['GeoAreaName'].unique():
    for year in df_long['Year'].unique():
        # 获取该国该年的所有性别数据
        country_year_data = df_long[
            (df_long['GeoAreaName'] == country) & 
            (df_long['Year'] == year)
        ]
        
        if len(country_year_data) == 0:
            continue
        
        # 策略1: 如果有BOTHSEX，优先使用
        bothsex_data = country_year_data[country_year_data['Sex'] == 'BOTHSEX']
        if len(bothsex_data) > 0:
            value = bothsex_data['Value'].iloc[0]
            result_list.append({
                'Country': country,
                'Year': year,
                'Value': value,
                'Source': 'BOTHSEX'
            })
            continue
        
        # 策略2: 如果有MALE和FEMALE，取平均
        male_data = country_year_data[country_year_data['Sex'] == 'MALE']
        female_data = country_year_data[country_year_data['Sex'] == 'FEMALE']
        
        if len(male_data) > 0 and len(female_data) > 0:
            male_value = male_data['Value'].iloc[0]
            female_value = female_data['Value'].iloc[0]
            avg_value = (male_value + female_value) / 2
            result_list.append({
                'Country': country,
                'Year': year,
                'Value': avg_value,
                'Source': 'MALE+FEMALE Average'
            })
            continue
        
        # 策略3: 只有一个性别，直接使用
        if len(country_year_data) == 1:
            value = country_year_data['Value'].iloc[0]
            sex = country_year_data['Sex'].iloc[0]
            result_list.append({
                'Country': country,
                'Year': year,
                'Value': value,
                'Source': f'Only {sex}'
            })

# 转换为DataFrame
df_result = pd.DataFrame(result_list)

# 添加SDG编号
df_result['SDG'] = 16

# 调整列顺序
df_result = df_result[['Country', 'Year', 'SDG', 'Value']]

# 排序
df_result = df_result.sort_values(['Country', 'Year'])

# 保存
df_result.to_csv('sdg16.csv', index=False, encoding='utf-8-sig')

print(f"\n 处理完成！共 {len(df_result)} 行数据")
print(df_result.head(10))
```
